```python
from easynmt import EasyNMT
```

```python
model = EasyNMT('opus-mt')#importing the model, for translation and translate for you, process of converting words into numbers
```

```python
sentences = [
    'Jesus is alive!',
    'Hands on projects is our priority',
    "I don't have anything",
    "It is a rainy day",
    "'It is what it is!': Legend B once said"
]
```

```python
translations_fr = model.translate(sentences, target_lang='fr')
translations_rw = model.translate(sentences, target_lang='rw')
```
